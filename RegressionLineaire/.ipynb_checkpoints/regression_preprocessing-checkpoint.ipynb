```python
#Import des données
working_directory = "/home/samy/csv_pickle_parquet/"

anjukedata = pd.read_pickle(working_directory + "utseus-anjuke-real-estate.pk")
poi = pd.read_csv(working_directory+"shanghaiPOI2017.csv",sep=";",index_col=0)
anjuke = pd.DataFrame(columns = anjukedata[0],data=anjukedata[1:])
```

```python

anjuke[['longitude', 'latitude', 'bedroom', 'room', 'surface',
       'price', 'onesquaremeter']] = anjuke[['longitude', 'latitude', 'bedroom', 'room', 'surface',
       'price', 'onesquaremeter']].apply(pd.to_numeric)
anjuke.dropna(inplace=True)

anjuke_2 = anjuke[["id","longitude","latitude","onesquaremeter"]]


anjuke_gpd = gpd.GeoDataFrame(
    anjuke_2,
    geometry = gpd.points_from_xy(anjuke_2.longitude,anjuke_2.latitude,crs=4326)
)


ymax,ymin,xmax,xmin = [31.514999,30.975,121.804611,121.10157]

anjuke_gpd = anjuke_gpd.cx[xmin:xmax,ymin:ymax]
anjuke_distance = anjuke_gpd[["id","longitude","latitude","geometry","onesquaremeter"]].to_crs(4576)

```

```python
subways = poi[poi["GBCODE"]==200402][["POIID","GCJX","GCJY","GPSX","GPSY"]]

for val in ["GCJX","GCJY","GPSX","GPSY"]:
    subways[val] = subways[val].str.replace(",",".")


    
subways["GCJX"] = subways["GCJX"].astype(float)
subways["GCJY"] = subways["GCJY"].astype(float)
subways["GPSX"] = subways["GPSX"].astype(float)
subways["GPSY"] = subways["GPSY"].astype(float)

subways_gpd = gpd.GeoDataFrame(
    subways,
    geometry=gpd.points_from_xy(subways.GPSX,subways.GPSY,crs=4326)
)

subways_gpd = subways_gpd.cx[xmin:xmax,ymin:ymax]
subways_distance = subways_gpd[["POIID","GPSX","GPSY","geometry"]].to_crs(4576)


subways_distance
```

```python
x = subways_gpd.geometry.x
y = subways_gpd.geometry.y
subway_points = np.array(list(zip(x,y)))

x = anjuke_gpd.geometry.x
y = anjuke_gpd.geometry.y
anjuke_points = np.array(list(zip(x,y)))


tree = BallTree(subway_points, leaf_size=2, metric='euclidean')
anjuke_gpd['nearest_sklearn'], _ = tree.query(anjuke_points, k=1)

linreg_dataset = anjuke_gpd[["nearest_sklearn","onesquaremeter"]]


price_threshold = linreg_dataset.onesquaremeter.quantile(0.5) + 3 * np.sqrt(linreg_dataset.onesquaremeter.var())

linreg_dataset = linreg_dataset[linreg_dataset["onesquaremeter"] < price_threshold]
```

```python
linreg_dataset
```

```python
price_mean = linreg_dataset.onesquaremeter.mean()
distance_mean = linreg_dataset.nearest_sklearn.mean()

dx_dy = (linreg_dataset["nearest_sklearn"] - distance_mean) * (linreg_dataset["onesquaremeter"] - price_mean)
dx_squared = (linreg_dataset["nearest_sklearn"] - distance_mean)**2

beta = dx_dy.sum() / dx_squared.sum()


alpha = price_mean - beta * distance_mean

beta
```
