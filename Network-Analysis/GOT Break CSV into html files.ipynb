```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
all_html = pd.read_csv("/home/samy/csv_pickle_parquet/awoiaf_raw_html_v2.csv")
```

```python
all_html
```

```python
for row in tqdm(range(len(all_html))):
    chemin_fichier = os.path.join("/home/samy/csv_pickle_parquet/html_got",all_html.loc[row,"character"].lower().replace(" ","_") + ".html")
    with open(chemin_fichier, "w") as fichier:
        fichier.write(all_html.loc[row,"raw_data"])
```
