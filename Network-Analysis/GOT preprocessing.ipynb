```python
import pandas as pd
from tqdm import tqdm

def splitter_15(text):
    return text.strip().split("\n\n\n\n\n\n")

def splitter_234(text,split_str):
    list = text.strip().split(split_str)
    output = []
    stop_value = 0
    while stop_value < len(list):
        output.append(list[stop_value] + list[stop_value + 1])
        stop_value += 2
    return output

def name_extractor(old_string):
    string = old_string.strip()
    if string[:28] != "Trueborn Lord of Winterfell.":
        name =""
        for i in range(len(string)):
            if string[i] != "\n" and string[i] != "\t":
                name += string[i]
            else:
                break
        return name
    else:
        return 'Trueborn Lord of Winterfell.'
```

```python
filenames = ["GOT1.txt","GOT2.txt","GOT3.txt","GOT4.txt","GOT5.txt"]

data_got = pd.DataFrame(columns = ["tome","nom_chapitre","contenu"])

for file in tqdm(filenames):
    with open(f"/home/samy/csv_pickle_parquet/{file}","r") as txt:
        current_file = txt.read()
        if file == "GOT1.txt" or file=="GOT5.txt":
            splitted_file = splitter_15(current_file)
        elif file=="GOT2.txt" or file=="GOT4.txt":
            splitted_file = splitter_234(current_file,"\n\n")
        elif file == "GOT3.txt":
            splitted_file = splitter_234(current_file,"\n\n\n")
        for chapter in splitted_file:
            data_got.loc[len(data_got),:] = [file[3],name_extractor(chapter),chapter]
```
